## 4.2.2 Solution

### 4.2.2.1 Import data and take a first look

```python
#import data
#just put all your csv's where you run the notebook from

boys = pd.read_csv(folder + 'data/boys.csv', index_col = 0)
girls = pd.read_csv(folder + 'data/girls.csv', index_col = 0)

#transform into np arrays
np_girls = np.array(girls)
np_boys = np.array(boys)
```

### 4.2.2.2 Perform the t-test

'''
unpaired two sample t-test;
H0: girls = boys
HA: boys != girls
'''

```python
#http://evafengeva.blogspot.be/2017/09/how-to-interpret-results-of-two-sample.html
t, p = stats.ttest_ind(np_boys, np_girls, axis = 0, equal_var=True)
```

```python
print(t)
print(p)
```

The p-value is 0.0000000000016574, which is very small. So we reject the null hypothesis. There is a significant difference between boys and girls.
